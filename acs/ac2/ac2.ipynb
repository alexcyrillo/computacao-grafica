# Exercício 1: Translação Simples

## Dado o ponto ( P(2, 3) ), aplique a translação com vetor ( (4, -2) ).

- Perguntas: Qual é o novo ponto ( P' )? Quais coordenadas foram alteradas?


```python
def plot_points(original_point, transformed_point, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(*original_point, color='blue', label=f"Original")
    plt.scatter(*transformed_point, color='red', label=f"Transformado")
    
    vetor = transformed_point - original_point
  
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

ponto = np.array([2, 3])

vetor_translacao = np.array([4, -2])
ponto_transformado = ponto + vetor_translacao

plot_points(ponto, ponto_transformado, "Exercício 1: Translação Simples")
```

# Exercício 2: Escala Uniforme
Considere um triângulo com vértices ( A(1, 1), B(3, 1) ) e ( C(2, 4) ). Aplique uma escala uniforme de fator 2.

- Perguntas: Quais são as novas coordenadas dos vértices após a transformação? O que acontece com o tamanho do triângulo?


```python
def plot_shapes(original_shape, transformed_shape, title):
    plt.figure(figsize=(8, 6))

    original_closed = np.vstack([original_shape, original_shape[0]])
    transformed_closed = np.vstack([transformed_shape, transformed_shape[0]])
    
    plt.plot(*original_closed.T, marker='o', label="Original")
    plt.plot(*transformed_closed.T, marker='o', label="Transformado", linestyle="--")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

triangulo = np.array([[1, 1], [3, 1], [2, 4]])

fator_escala = 2
triangulo_transformado = triangulo * fator_escala

plot_shapes(triangulo, triangulo_transformado, "Exercício 2: Escala Uniforme")
```

# Exercício 4: Rotação em Torno da Origem

## Rotacione o ponto ( P(1, 0) ) em 90° no sentido anti-horário em torno da origem.

- Pergunta: Qual é a nova posição do ponto ( P' ) após a rotação?


```python
def plot_points(original_point, transformed_point, title):
    plt.figure(figsize=(6, 6))
    plt.scatter(*original_point, color='blue', label=f"Original")
    plt.scatter(*transformed_point, color='red', label=f"Transformado")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

ponto = np.array([1, 0])

angulo_rad = np.radians(90)
matriz_rotacao = np.array([[np.cos(angulo_rad), -np.sin(angulo_rad)],
                               [np.sin(angulo_rad), np.cos(angulo_rad)]])
ponto_transformado = np.dot(matriz_rotacao, ponto)

plot_points(ponto, ponto_transformado, "Exercício 4: Rotação em Torno da Origem")
```

# Exercício 5: Rotação de um Polígono

## Um quadrado tem vértices ( A(1, 1), B(1, 4), C(4, 4), D(4, 1) ). Aplique uma rotação de 45° no sentido horário.

- Pergunta: Quais são as novas coordenadas dos vértices?


```python
def plot_shapes(original_shape, transformed_shape, title):
    plt.figure(figsize=(8, 6))

    original_closed = np.vstack([original_shape, original_shape[0]])
    transformed_closed = np.vstack([transformed_shape, transformed_shape[0]])
    
    plt.plot(*original_closed.T, marker='o', label="Original")
    plt.plot(*transformed_closed.T, marker='o', label="Transformado", linestyle="--")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

quadrado = np.array([[1, 1], [1, 4], [4, 4], [4, 1]])

angulo_rad = np.radians(-45)
matriz_rotacao = np.array([[np.cos(angulo_rad), -np.sin(angulo_rad)],
                               [np.sin(angulo_rad), np.cos(angulo_rad)]])
quadrado_transformado = np.dot(quadrado, matriz_rotacao.T)

plot_shapes(quadrado, quadrado_transformado, "Exercício 5: Rotação de um Polígono")
```

# Exercício 9: Composição de Transformações

## Dado o ponto ( P(3, 2) ), aplique as seguintes transformações consecutivas:

- 1. Uma translação com vetor ( (1, -1) ).

- 2. Uma rotação de 90° no sentido anti-horário.

- 3. Uma escala uniforme com fator 2.

- Pergunta: Qual é a nova posição do ponto ( P' ) após todas as transformações?


```python
def plot_shapes(original_shape, transformed_shape, title):
    plt.figure(figsize=(8, 6))
    original_closed = np.vstack([original_shape, original_shape[0]])
    transformed_closed = np.vstack([transformed_shape, transformed_shape[0]])
    
    plt.plot(*original_closed.T, marker='o', label="Original")
    plt.plot(*transformed_closed.T, marker='o', label="Transformado", linestyle="--")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

ponto_inicial = np.array([3, 2])

vetor_translacao = np.array([1, -1])
ponto_passo1 = ponto_inicial + vetor_translacao

angulo_rad = np.radians(90)
matriz_rotacao = np.array([[np.cos(angulo_rad), -np.sin(angulo_rad)],
                               [np.sin(angulo_rad), np.cos(angulo_rad)]])
ponto_passo2 = np.dot(matriz_rotacao, ponto_passo1)

fator_escala = 2
ponto_final = ponto_passo2 * fator_escala

plt.figure(figsize=(8, 8))
plt.scatter(*ponto_inicial, s=100, color='blue', label=f'Original')
plt.scatter(*ponto_passo1, s=100, color='orange', label=f'1. Translação')
plt.scatter(*ponto_passo2, s=100, color='green', label=f'2. Rotação')
plt.scatter(*ponto_final, s=100, color='red', label=f'3. Escala')
plt.title("Exercício 9: Composição de Transformações")
plt.legend()
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.axis('equal')
plt.show()
```

# Exercício 10: Combinação de Transformações em uma Figura

## Dado um retângulo com vértices ( A(1, 1), B(5, 1), C(5, 3), D(1, 3) ), aplique as seguintes transformações em sequência:

- 1. Translação com vetor ( (-2, 3) ).

- 2. Escala não uniforme com fatores ( 1.5 ) no eixo (x) e (0.5) no eixo (y).

- 3. Reflexão em relação ao eixo (y).

- Pergunta: Quais são as novas coordenadas dos vértices após as três transformações?


```python
def plot_shapes(original_shape, transformed_shape, title):
    plt.figure(figsize=(8, 6))

    original_closed = np.vstack([original_shape, original_shape[0]])
    transformed_closed = np.vstack([transformed_shape, transformed_shape[0]])
    
    plt.plot(*original_closed.T, marker='o', label="Original")
    plt.plot(*transformed_closed.T, marker='o', label="Transformado", linestyle="--")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

retangulo_inicial = np.array([[1, 1], [5, 1], [5, 3], [1, 3]])

vetor_translacao = np.array([-2, 3])
retangulo_passo1 = retangulo_inicial + vetor_translacao

fatores_escala = np.array([1.5, 0.5])
retangulo_passo2 = retangulo_passo1 * fatores_escala

matriz_reflexao = np.array([[-1, 0], [0, 1]])
retangulo_final = np.dot(retangulo_passo2, matriz_reflexao)

plot_shapes(retangulo_inicial, retangulo_final, "Exercício 10: Combinação de Transformações")

plt.figure(figsize=(10, 8))
original_closed = np.vstack([retangulo_inicial, retangulo_inicial[0]])
passo1_closed = np.vstack([retangulo_passo1, retangulo_passo1[0]])
passo2_closed = np.vstack([retangulo_passo2, retangulo_passo2[0]])
final_closed = np.vstack([retangulo_final, retangulo_final[0]])

plt.plot(*original_closed.T, marker='o', label='Original')
plt.plot(*passo1_closed.T, marker='o', label='1. Translação', linestyle=':')
plt.plot(*passo2_closed.T, marker='o', label='2. Escala', linestyle='-.')
plt.plot(*final_closed.T, marker='o', label='3. Reflexão (Final)', linestyle='--')
plt.title("Exercício 10: Visualização dos Passos da Transformação")
plt.legend()
plt.grid(True)
plt.show()
```
